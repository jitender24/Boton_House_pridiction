```python
import numpy as np
```

```python
def cost(x, y, m):
    yPred = np.dot(x, m)
    return ((yPred - y)**2).mean()
```

```python
def calc_func(i, x, m):
    k, n = x.shape
    func = 0
    for j in range(n):
        func += x[i][j] * m[j]
    return func
```

```python
def step_gradient(x,y, learning_rate, m):
    k, n = x.shape
    slope_m = np.array([0 for i in range(n)], dtype = 'float32')
    for j in range(n):
        for i in range(k):
            slope_m[j] += (2/k)*(calc_func(i, x, m) - y[i])*x[i][j]
    
    for j in range(n):
        m[j] = m[j] - learning_rate * slope_m[j]
    return m
```

```python
def gradient_descent(x,y,learning_rate, num_iter):
    k, n = x.shape
    print(k,n)
    m = np.array([0 for i in range(n)], dtype = 'float32')
    print(m)
    
    for i in range(num_iter):
        m = step_gradient(x,y, learning_rate,m)
        c= cost(x,y,m)
        print(i,":", c)
    return m
```
